```python
data = train1
```

```python
data.dtypes
```

```python
#create one hot encoding 
cat_vars=['Pclass','Sex','Cabin','Embarked']
for var in cat_vars:
    cat_list='var'+'_'+var
    cat_list = pd.get_dummies(data[var], prefix=var)
    data1=data.join(cat_list)
    data=data1
cat_vars=['Pclass','Sex','Cabin','Embarked']
data_vars=data.columns.values.tolist()
to_keep=[i for i in data_vars if i not in cat_vars]
```

```python
data_final=data[to_keep]
data_final.columns.values
```

```python
data_final_vars=data_final.columns.values.tolist()
y=['Survived']
X=[i for i in data_final_vars if i not in ["Name","Survived","PassengerId","Ticket","Fam"]]
```

```python
len(X)
```

```python
data_final[X].dtypes
```

```python
from sklearn import datasets
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
rfe = RFE(logreg, 10)
rfe = rfe.fit(data_final[X], data_final[y] )
print(rfe.support_)
print(rfe.ranking_)
```

```python
data_final[X].dtypes
```

```python
cols = ["Pclass_2","Pclass_3","Sex_female","Sex_male","Cabin_A","Cabin_E","Cabin_F","Cabin_G","Cabin_T","Embarked_S"]
X = data_final[cols]
y = data_final["Survived"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
y_pred = logreg.predict(X_test)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logreg.score(X_test, y_test)))
```

```python
from sklearn import model_selection
from sklearn.model_selection import cross_val_score
kfold = model_selection.KFold(n_splits=10, random_state=7)
modelCV = LogisticRegression()
scoring = 'accuracy'
results = model_selection.cross_val_score(modelCV, X_train, y_train, cv=kfold, scoring=scoring)
print("10-fold cross validation average accuracy: %.3f" % (results.mean()))
```
